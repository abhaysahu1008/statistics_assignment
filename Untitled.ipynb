```python
#1. Generate a list of 100 integers containing values between 90 to 130 and store it in the variable "int_list". After generating the list, find the following:

import random
import math
from collections import Counter

# Generate the list of integers
int_list = [random.randint(90, 130) for _ in range(100)]

# Functions for mean and median
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    if n % 2 == 0:
        median = (sorted_numbers[n//2 - 1] + sorted_numbers[n//2]) / 2
    else:
        median = sorted_numbers[n//2]
    return median

# Function for mode
def calculate_mode(numbers):
    counter = Counter(numbers)
    max_count = max(counter.values())
    modes = [k for k, v in counter.items() if v == max_count]
    return modes

# Function for weighted mean
def calculate_weighted_mean(values, weights):
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    total_weight = sum(weights)
    return weighted_sum / total_weight

# Function for geometric mean
def calculate_geometric_mean(numbers):
    product = math.prod(numbers)
    return product ** (1 / len(numbers))

# Function for harmonic mean
def calculate_harmonic_mean(numbers):
    return len(numbers) / sum(1 / x for x in numbers)

# Function for midrange
def calculate_midrange(numbers):
    return (min(numbers) + max(numbers)) / 2

# Function for trimmed mean
def calculate_trimmed_mean(numbers, percentage):
    sorted_numbers = sorted(numbers)
    n = len(numbers)
    trim_count = int(n * percentage)
    trimmed_numbers = sorted_numbers[trim_count:-trim_count]
    return sum(trimmed_numbers) / len(trimmed_numbers)

# Example usage
mean = calculate_mean(int_list)
median = calculate_median(int_list)
mode = calculate_mode(int_list)
weighted_mean = calculate_weighted_mean(int_list, [random.random() for _ in range(100)])
geometric_mean = calculate_geometric_mean([num for num in int_list if num > 0])  # Ensure all numbers are positive
harmonic_mean = calculate_harmonic_mean(int_list)
midrange = calculate_midrange(int_list)
trimmed_mean = calculate_trimmed_mean(int_list, 0.1)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Weighted Mean: {weighted_mean}")
print(f"Geometric Mean: {geometric_mean}")
print(f"Harmonic Mean: {harmonic_mean}")
print(f"Midrange: {midrange}")
print(f"Trimmed Mean: {trimmed_mean}")

```

```python
#2. Generate a list of 500 integers containing values between 200 to 300 and store it in the variable "int_list2". After generating the list, find the following:


import random
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate the list of integers
int_list2 = [random.randint(200, 300) for _ in range(500)]

# Function for visualizations
def plot_visualizations(data):
    plt.figure(figsize=(18, 6))
    
    # Frequency & Gaussian distribution
    plt.subplot(1, 3, 1)
    sns.histplot(data, kde=False, stat="density", bins=30, color='blue')
    sns.kdeplot(data, color='red')
    plt.title("Frequency & Gaussian distribution")

    # Frequency smoothened KDE plot
    plt.subplot(1, 3, 2)
    sns.histplot(data, kde=True, stat="density", bins=30, color='blue')
    plt.title("Frequency smoothened KDE plot")

    # Gaussian distribution & smoothened KDE plot
    plt.subplot(1, 3, 3)
    sns.kdeplot(data, color='red')
    plt.title("Gaussian distribution & smoothened KDE plot")
    
    plt.tight_layout()
    plt.show()

plot_visualizations(int_list2)

# Functions for statistical calculations
def calculate_range(numbers):
    return max(numbers) - min(numbers)

def calculate_variance(numbers):
    mean = sum(numbers) / len(numbers)
    return sum((x - mean) ** 2 for x in numbers) / len(numbers)

def calculate_std_dev(numbers):
    variance = calculate_variance(numbers)
    return variance ** 0.5

def calculate_iqr(numbers):
    sorted_numbers = sorted(numbers)
    q1 = np.percentile(sorted_numbers, 25)
    q3 = np.percentile(sorted_numbers, 75)
    return q3 - q1

def calculate_coefficient_of_variation(numbers):
    mean = sum(numbers) / len(numbers)
    std_dev = calculate_std_dev(numbers)
    return std_dev / mean

def calculate_mad(numbers):
    mean = sum(numbers) / len(numbers)
    return sum(abs(x - mean) for x in numbers) / len(numbers)

def calculate_quartile_deviation(numbers):
    sorted_numbers = sorted(numbers)
    q1 = np.percentile(sorted_numbers, 25)
    q3 = np.percentile(sorted_numbers, 75)
    return (q3 - q1) / 2

def calculate_range_based_coefficient_of_dispersion(numbers):
    range_ = calculate_range(numbers)
    mean = sum(numbers) / len(numbers)
    return range_ / mean

# Example usage
range_ = calculate_range(int_list2)
variance = calculate_variance(int_list2)
std_dev = calculate_std_dev(int_list2)
iqr = calculate_iqr(int_list2)
coefficient_of_variation = calculate_coefficient_of_variation(int_list2)
mad = calculate_mad(int_list2)
quartile_deviation = calculate_quartile_deviation(int_list2)
range_based_coefficient_of_dispersion = calculate_range_based_coefficient_of_dispersion(int_list2)

print(f"Range: {range_}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"IQR: {iqr}")
print(f"Coefficient of Variation: {coefficient_of_variation}")
print(f"MAD: {mad}")
print(f"Quartile Deviation: {quartile_deviation}")
print(f"Range-Based Coefficient of Dispersion: {range_based_coefficient_of_dispersion}")
```

```python
#3. Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("Values and probabilities must be the same length.")
        if not np.isclose(sum(probabilities), 1):
            raise ValueError("Probabilities must sum to 1.")
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        expected_val = self.expected_value()
        return sum(prob * (value - expected_val) ** 2 for value, prob in zip(self.values, self.probabilities))

# Example usage
values = [1, 2, 3, 4, 5, 6]
probabilities = [1/6, 1/6, 1/6, 1/6, 1/6, 1/6]  # Fair six-sided die
drv = DiscreteRandomVariable(values, probabilities)
print(f"Expected Value: {drv.expected_value()}")
print(f"Variance: {drv.variance()}")

```

```python
#4. Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and variance of the outcomes.

import random

def simulate_die_rolls(n_rolls):
    outcomes = [random.randint(1, 6) for _ in range(n_rolls)]
    return outcomes

def calculate_expected_value(outcomes):
    return sum(outcomes) / len(outcomes)

def calculate_variance(outcomes):
    mean = calculate_expected_value(outcomes)
    return sum((x - mean) ** 2 for x in outcomes) / len(outcomes)

# Example usage
n_rolls = 10000
outcomes = simulate_die_rolls(n_rolls)
expected_value = calculate_expected_value(outcomes)
variance = calculate_variance(outcomes)
print(f"Expected Value: {expected_value}")
print(f"Variance: {variance}")
```

```python
#5. Create a Python function to generate random samples from a given probability distribution (e.g., binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_random_samples(distribution, params, n_samples):
    if distribution == 'binomial':
        samples = np.random.binomial(params['n'], params['p'], n_samples)
    elif distribution == 'poisson':
        samples = np.random.poisson(params['lambda'], n_samples)
    else:
        raise ValueError("Unsupported distribution")

    mean = np.mean(samples)
    variance = np.var(samples)
    return samples, mean, variance

# Example usage
binomial_params = {'n': 10, 'p': 0.5}
poisson_params = {'lambda': 5}

binomial_samples, binomial_mean, binomial_variance = generate_random_samples('binomial', binomial_params, 10000)
poisson_samples, poisson_mean, poisson_variance = generate_random_samples('poisson', poisson_params, 10000)

print(f"Binomial - Mean: {binomial_mean}, Variance: {binomial_variance}")
print(f"Poisson - Mean: {poisson_mean}, Variance: {poisson_variance}")
```

```python
#6. Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute the mean, variance, and standard deviation of the samples.

def generate_gaussian_samples(mean, std_dev, n_samples):
    samples = np.random.normal(mean, std_dev, n_samples)
    calculated_mean = np.mean(samples)
    variance = np.var(samples)
    std_dev_calculated = np.std(samples)
    return samples, calculated_mean, variance, std_dev_calculated

# Example usage
mean = 0
std_dev = 1
n_samples = 10000

samples, calculated_mean, variance, std_dev_calculated = generate_gaussian_samples(mean, std_dev, n_samples)

print(f"Mean: {calculated_mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev_calculated}")
```

```python
#8.Write a Python function to calculate the probability density function (PDF) of a continuous random
#variable for a given normal distribution.

import math

def normal_pdf(x, mean, std_dev):
    coefficient = 1 / (std_dev * math.sqrt(2 * math.pi))
    exponent = math.exp(-((x - mean) ** 2) / (2 * std_dev ** 2))
    return coefficient * exponent

# Example usage
mean = 0
std_dev = 1
x = 0

pdf_value = normal_pdf(x, mean, std_dev)
print(f"PDF value at x={x} for N({mean}, {std_dev}^2): {pdf_value}")
```

```python
#9. Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

import math

def exponential_cdf(x, rate):
    if x < 0:
        return 0.0
    return 1 - math.exp(-rate * x)

# Example usage
rate = 0.5  # lambda
x = 2

cdf_value = exponential_cdf(x, rate)
print(f"CDF value at x={x} for exponential distribution with rate={rate}: {cdf_value}")
```

```python
#10. Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

import math

def poisson_pmf(k, rate):
    return (rate ** k * math.exp(-rate)) / math.factorial(k)

# Example usage
rate = 3  # lambda
k = 2

pmf_value = poisson_pmf(k, rate)
print(f"PMF value at k={k} for Poisson distribution with rate={rate}: {pmf_value}")
```

```python
'''11. A company wants to test in a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.


To generate the data use the following command:

python

import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors  

new_layout = np.array([1] * 70 + [0] * 930)


Apply z-test to find which layout is successful'''

import numpy as np
import scipy.stats as stats

# Generate the data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Conversion rates
conversion_old = np.mean(old_layout)
conversion_new = np.mean(new_layout)

# Number of visitors
n_old = len(old_layout)
n_new = len(new_layout)

# Pooled conversion rate
pooled_conversion = (conversion_old * n_old + conversion_new * n_new) / (n_old + n_new)

# Standard error
se = np.sqrt(pooled_conversion * (1 - pooled_conversion) * (1/n_old + 1/n_new))

# Z-score
z = (conversion_new - conversion_old) / se

# P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z)))  # Two-tailed test

# Significance level
alpha = 0.05

# Decision
if p_value < alpha:
    print(f"Reject the null hypothesis. P-value: {p_value:.4f}. Z-score: {z:.4f}")
    print("The new layout has a significantly different conversion rate compared to the old layout.")
else:
    print(f"Fail to reject the null hypothesis. P-value: {p_value:.4f}. Z-score: {z:.4f}")
    print("There is no significant difference in conversion rates between the new and old layouts.")


```
